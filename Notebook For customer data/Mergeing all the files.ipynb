# Lets make a list of the files paths

```python
file_paths =['../Dataset/raw/Oct 1, 2023 - Oct 15, 2023.csv','../Dataset/raw/Oct 15, 2023 - Oct 29, 2023.csv','../Dataset/raw/Oct 29, 2023 - Nov 12, 2023.csv',
             '../Dataset/raw/Nov 12, 2023 - Nov 26, 2023.csv','../Dataset/raw/Nov 26, 2023 - Dec 10, 2023.csv','../Dataset/raw/Dec 10, 2023 - Dec 24, 2023.csv',
             '../Dataset/raw/Dec 24, 2023 - Jan 7, 2024.csv','../Dataset/raw/Jan 7, 2024 - Jan 21, 2024.csv','../Dataset/raw/Jan 21, 2024 - Feb 4, 2024.csv',
             '../Dataset/raw/Feb 4, 2024 - Feb 18, 2024.csv','../Dataset/raw/Feb 18, 2024 - Mar 3, 2024.csv','../Dataset/raw/Mar 3, 2024 - Mar 17, 2024.csv',
             '../Dataset/raw/Mar 17, 2024 - Mar 31, 2024.csv','../Dataset/raw/Mar 31, 2024 - Apr 14, 2024.csv','../Dataset/raw/Apr 14, 2024 - Apr 28, 2024.csv']
```

```python
print(file_paths)
```

```python
dataframes = [pd.read_csv(file,delimiter =';') for file in file_paths]
```

```python
combined_dataframes = pd.concat(dataframes, ignore_index = True)
```

```python
combined_dataframes
```

```python
# here i have removed the duplicated data from the rows to avoid duplication duirng the combined of diffrenct csv file 

remdup_combined_dataframes =combined_dataframes.drop_duplicates()
```

```python
remdup_combined_dataframes.shape
```

```python
# After merging and removing duplicate file 
# Lets export new csv file having name Oct 1,2023 - Apr 28, 2024.csv in the preprocessed directory
csv_filename = "../Dataset/preprocessed/Oct 1,2023 - Apr 28, 2024_Merge.csv"
remdup_combined_dataframes.to_csv(csv_filename,index= False,)
print("saved")
```
